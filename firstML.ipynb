This model provides information about the shows people like based on their age and gender.

The "fav_data" file has 3 columns:

```| Gender | Age | FavShow |```

- **Gender**: Male is represented as 1, and female as 0.
- **Age**: Contains the person's age.
- **FavShow**: Contains the shows (movies, animes, news, etc.) they like.

## Importing modules

````python
from pandas import read_csv # to load data
from sklearn.tree  import DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder

###### Pat 2 - MODEL QUALITY
# we take the absolute value of each error.
#  (error = actualData − predictedData)
# This converts each error to a positive number.
# We then take the average of those absolute errors.
# This is our measure of model quality
#### adding mean_absolute_error
# from sklearn.metrics import mean_absolute_error
````

## Preparing Data for training

````python
def createFile(name):
  with open(name, 'w') as f:
    content = '''Gender,Age,FavShow
1,10,kidsShow
0,10,cuteShow
1,15,cartoons
0,15,cartoons
1,20,Animes
0,20,drama
1,25,Movies
0,25,drama
1,30,news
0,30,drama'''
    f.write(content)

try:
  data = read_csv('fav_data.csv')
except:
  createFile('fav_data.csv')
  data = read_csv('fav_data.csv')
data
````

### Output data labeling

````python
y = data.FavShow

# decisionTreeRegressor only work on the float,
# str cant convert to the float here so, we are labeling it
le = LabelEncoder()
y_encoded = le.fit_transform(y)

for i in range(len(y_encoded)):
    print(f"{y_encoded[i]} = {y[i]}")
````

### Getting Features

````python
fav_show_features = ['Gender', 'Age']
X = data[fav_show_features]
X
````

## Defining and training model

````python
# defining and training model
model = DecisionTreeRegressor(random_state=1)
model.fit(X, y_encoded)
````

## Preparing Testing Data

````python
# test data prepare
# change the data as you like
def createFile(name):
  with open(name, 'w') as f:
    content = '''Gender,Age
1,11
0,12
1,40
0,35
1,5
'''
    f.write(content)

try:
  test_data = read_csv('test_data.csv')
except:
  createFile('test_data.csv')
  test_data = read_csv('test_data.csv')
test_data
````

## Final Prediction

````python
# prediction
model.predict(test_data)
````
